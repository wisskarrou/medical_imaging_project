```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_mean_roc_auc(model,title=''):
    df = model
    df['mean_roc_auc'] = (df['val_roc_auc_C_1']+df['val_roc_auc_C_2']+df['val_roc_auc_C_5'])/3
    df[['mean_roc_auc','val_roc_auc_C_1','val_roc_auc_C_2','val_roc_auc_C_5']].plot()
    print(f"Maximal mean ROC AUC = {max((model['val_roc_auc_C_1']+model['val_roc_auc_C_2']+model['val_roc_auc_C_5'])/3)}")
    print(f"Reached after {np.argmax(model['val_roc_auc_C_1']+model['val_roc_auc_C_2']+model['val_roc_auc_C_5'])} epochs")

def plot_val_loss(model,title=''):
    df = model
    df[['val_loss_C_1','val_loss_C_2','val_loss_C_5']].plot()

def plot_losses(model,val_center):
  model[['val_loss_'+val_center,'train_loss_'+val_center]].plot()
```

```python
model = pd.read_csv("domain_adaptation_v2_fc256_gat256_fc128_fc64_fc1_batch8_lr1eminus5_epochs40_tanh_no_dropout.csv")
```

```python
plot_mean_roc_auc(model)
```

```python
plot_val_loss(model)
```
